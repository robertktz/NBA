```python
#changing the game_date format
df.game_date = pd.to_datetime(df.game_date)
df.head(1)
```

Dropping Seasons with smaller sample sizes (keep all above 1000 observations)

```python
df.season_id.value_counts()

#dropping the season if value count is less than 1000

df = df[df.season_id.value_counts()[df.season_id].values > 1000]
df.season_id.value_counts() #checking if the code worked

```

```python
#checking which season_id belongs to which season

df.season_id.value_counts().index

# Assuming the index is stored in the variable 'index'
index = pd.Index([22005, 22004, 22016, 22015, 22014, 22013, 22010, 22009, 22008,
                  22007, 22006, 22021, 22018, 22012, 22002, 22001, 22000, 21999,
                  21995, 22003, 21997, 21996, 21994, 21992, 21993, 21991, 22020,
                  21989, 21990, 22019, 22022], dtype='int64')

# Create an empty dictionary
season_dict = {}

# Iterate over the index and create key-value pairs in the dictionary
for season_id in index:
    season_start = str(season_id)[-2:]  # Extract the last 2 digits as the start year
    season_end = str(int(season_start) + 1).zfill(2)  # Calculate the end year as start year + 1
    season_year = f"{season_start}-{season_end}"  # Create the season year in the format "start-end"
    season_dict[season_id] = season_year

# Print the dictionary
print(season_dict)
```

```python
# Map the values of the season_id column using the dictionary and putting them next to season_id

df['season'] = df.season_id.map(season_dict)
df.head(1)
```

replacing the season_id column with the new 'season' column

```python
# Remove the "season_id" column
df.drop("season_id", axis=1, inplace=True)

# Get the column names of the DataFrame
columns = df.columns.tolist()

# Insert the "season" column at the first position
columns.insert(0, "season")

# Reorder the DataFrame with the new column order
df = df[columns]
df.head(1)
```

```python
# drop season as we have now 2
# by dropping the first column it should do

df = df.iloc[:, 1:]
```

I want to change the order of the columns, to have everything much clearer...

```python
desired_order = ['season', 'game_date','wl_home', 'team_name_home', 'team_name_away','min',
                 'fgm_home', 'fgm_away', 'fga_home', 'fga_away', 'fg_pct_home', 'fg_pct_away',
                 'fg3m_home', 'fg3m_away', 'fg3a_home', 'fg3a_away', 'fg3_pct_home', 'fg3_pct_away',
                 'ftm_home', 'ftm_away', 'fta_home', 'fta_away', 'ft_pct_home', 'ft_pct_away',
                 'oreb_home', 'oreb_away', 'dreb_home', 'dreb_away', 'reb_home', 'reb_away',
                 'ast_home', 'ast_away', 'stl_home', 'stl_away', 'blk_home', 'blk_away',
                 'tov_home', 'tov_away', 'pf_home', 'pf_away'
                 ]

df = df[desired_order]
df.head(1)
```

```python
# Preprocess the season column
df['season_start_year'] = df['season'].str.split('-').str[0].astype(int)

# Preprocess the game_date column
df['game_date'] = pd.to_datetime(df['game_date'])
df['game_year'] = df['game_date'].dt.year
df['game_month'] = df['game_date'].dt.month
df['game_day'] = df['game_date'].dt.day
```

```python
# Convert 'wl_home' to a numerical variable
df['wl_home'] = df['wl_home'].map({'W': 1, 'L': 0})
```
